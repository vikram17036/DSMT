
Model Interpretability

Abstract: This research investigates the prediction of chronic kidney disease (CKD) using a dataset collected over a 2-month period in India, encompassing 25 features related to various medical parameters such as blood pressure, blood glucose levels, and blood cell counts. Employing linear regression, decision tree regression, and Automated Machine Learning (AutoML), we seek to identify the most effective predictive model. Through a comprehensive analysis of regression coefficients, decision tree nodes, and the best-performing AutoML model, we aim to elucidate the interpretability of these models. Additionally, we conduct SHAP (SHapley Additive exPlanations) analysis to delve into the contribution of each feature across models, providing valuable insights into feature importance and model behavior. By combining traditional and automated modeling techniques with advanced interpretability analysis, our study offers a holistic understanding of the predictive factors associated with CKD, thus contributing to enhanced diagnosis and management of the disease.

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
import shap

numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X_train.select_dtypes(include=['object']).columns

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

logistic_model = LogisticRegression(max_iter=1000)

pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', logistic_model)])
pipeline.fit(X_train, y_train)

logistic_model = pipeline.named_steps['model']
coefficients = logistic_model.coef_[0]

X_train_numeric = pd.get_dummies(X_train)
X_test_numeric = pd.get_dummies(X_test)

X_test_numeric = X_test_numeric.reindex(columns=X_train_numeric.columns, fill_value=0)

explainer = shap.Explainer(logistic_model, X_train_numeric)
shap_values = explainer.shap_values(X_test_numeric)
```

```python
import shap

shap.summary_plot(shap_values, X_test_numeric)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import shap

numeric_transformer = SimpleImputer(strategy='mean')

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

tree_model = DecisionTreeClassifier(random_state=0)
tree_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', tree_model)])

tree_pipeline.fit(X_train, y_train)

explainer = shap.TreeExplainer(tree_pipeline.named_steps['model'])

X_test_preprocessed = preprocessor.transform(X_test)

X_test_preprocessed_dense = X_test_preprocessed.toarray()

shap_values = explainer.shap_values(X_test_preprocessed_dense)

shap.summary_plot(shap_values, X_test_preprocessed_dense)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plot_tree(tree_pipeline.named_steps['model'],
          feature_names=tree_pipeline.named_steps['preprocessor'].transformers_[0][2] +
                        list(tree_pipeline.named_steps['preprocessor'].named_transformers_['cat']['onehot'].get_feature_names_out()),
          class_names=tree_pipeline.named_steps['model'].classes_,
          filled=True)
plt.show()
```

Interpreting SHAP dependence plot for Linear and Tree-based model

Interpreting SHAP dependence plots for both linear and tree-based models provides insights into how the value of a specific feature influences the prediction of the model, either globally (across all data points) or for individual predictions. Here’s how to interpret these plots for both types of models:

Linear Model
For a linear model, SHAP dependence plots can be quite straightforward to interpret because the relationship between features and the target variable is linear by definition. The plot will typically show a straight line indicating that as the feature value increases or decreases, its impact on the prediction linearly increases or decreases in proportion.

Slope Direction: The direction of the slope (positive or negative) tells you whether the relationship between the feature and the target is positive (as the feature value increases, so does the prediction) or negative (as the feature value increases, the prediction decreases).
Slope Steepness: The steepness of the slope indicates the strength of the relationship. Steeper slopes mean the feature has a stronger influence on the prediction.
Divergence from Linearity: While linear models assume a linear relationship, SHAP values can sometimes reveal non-linear patterns if they are present due to interactions with other features. This can be an indication that the linear model is not perfectly capturing the relationship between the feature and the target.
Tree-based Model
For tree-based models, which can capture non-linear relationships and interactions between features, SHAP dependence plots can reveal more complex patterns:

Curve Shape: Unlike the linear model, the plot for a tree-based model might show curves, indicating non-linear relationships. For example, a U-shaped curve suggests that both high and low values of the feature strongly influence the prediction in the same direction, with a weaker influence from mid-range values.
Vertical Dispersion: We may see a lot of vertical dispersion at a given feature value, indicating variability in the SHAP value for the same feature value. This often points to interactions with other features, meaning the impact of the feature on the prediction changes depending on the values of other features.
Color Coding: SHAP plots often use color to represent the value of another feature, highlighting interactions. For instance, if points are colored by a second feature’s value, and you see a pattern in how colors are distributed across the plot (e.g., all high values of another feature cluster in a specific region of the plot), this indicates an interaction effect between the two features.
General Interpretation Tips
Feature Importance: In both linear and tree-based models, features with dependence plots showing larger deviations from the baseline prediction are generally more important to the model.

Interactions: For tree-based models, SHAP can explicitly show feature interactions in the dependence plot through color coding. This is less straightforward for linear models, where interactions need to be manually specified in the model.

Predictive Insight: Dependence plots can provide actionable insights. For instance, identifying thresholds where the impact of a feature on the prediction significantly changes can inform decision-making or highlight areas for further investigation.

SHAP dependence plots offer a powerful tool for understanding the relationship between features and predictions in both linear and tree-based models. By carefully analyzing these plots, you can gain deep insights into the behavior of your models and the underlying data.

Nodes: Each rectangle represents a node. The top node is called the root node, and it is the starting point of the decision-making process. Each node tests a specific attribute of the data.

Edges/branches: These are the lines connecting the nodes. They represent the outcome of the test at the previous node, leading to the next node.

Splitting criteria (features): At the top of each node, you will see a condition (e.g., num_sod <= 0.613). This is the decision rule that splits the data based on the value of a feature. num_ likely indicates a numerical feature, and cat_ indicates a categorical feature that has been numerically encoded.

Gini index: Below the splitting criteria is the Gini index (e.g., gini = 0.073). The Gini index is a measure of impurity or purity used to decide how to split the data. A Gini index of 0 indicates perfect purity, meaning all instances in the node belong to a single class.

Samples: The samples value (e.g., samples = 186) indicates the number of instances from the training dataset that fall into that node.

Value: The value array (e.g., value = [179, 1, 6]) shows the distribution of instances among classes in that node. Each number corresponds to the count of instances for a class. In a binary classification problem, there would typically be two numbers.

Class: The class label (e.g., class = ckd) indicates the dominant class in that node. The decision tree will predict the dominant class for instances that end up in this node.

Leaf nodes: The bottom nodes (e.g., gini = 0.0, samples = 3, value = [3, 0, 0], class = ckd) are called leaves or terminal nodes. They do not split further and represent the final decision of the tree.

Color coding: The colors likely represent the dominant class for visualization purposes. The intensity of the color might reflect the Gini index or the proportion of the majority class in the node.

Coefficient: This column represents the coefficient values associated with each feature in the model. Positive coefficients indicate that an increase in the corresponding feature value leads to an increase in the predicted outcome, while negative coefficients indicate the opposite.

Feature name: The feature names are broken down into two parts:

num__: This prefix indicates numerical features. For example, num__sc represents the feature "sc" (possibly serum creatinine) from the numerical features. cat__: This prefix indicates categorical features. For example, cat__pe_yes represents the presence of "pe" (possibly pedal edema) as a categorical feature where "yes" indicates its presence. Coefficient value: Each coefficient value represents the weight assigned to the corresponding feature by the model during training. Higher absolute values indicate a stronger influence of that feature on the model's predictions.

Based on the provided coefficients, we can infer the following:

Features with positive coefficients (num__sc, num__bu, num__al, cat__pe_yes, num__bp, etc.) have a positive impact on the predicted outcome. For example, as the values of these features increase, the likelihood of the predicted outcome also increases.

Features with negative coefficients (cat__dm_no, cat__pe_no, cat__wc_9500, num__sg, num__hemo, etc.) have a negative impact on the predicted outcome. For example, as the values of these features increase, the likelihood of the predicted outcome decreases.

The magnitude of the coefficients also provides information about the relative importance of each feature. Features with larger coefficient magnitudes have a stronger influence on the model's predictions compared to those with smaller magnitudes.
